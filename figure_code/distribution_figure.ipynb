```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import scipy
```

```python
x_min = 0.0
x_max = 10.0

mean1 = 5.0
mean2 = 5.3
mean3 = 5.5
mean5= 4.3
mean6 = 4.95
std_big = 1.1
std = .9
std_small = .7

x = np.linspace(x_min, x_max, 100)
y1 = scipy.stats.norm.pdf(x,mean1,std)
y2 = scipy.stats.norm.pdf(x,mean2,std)
y3 = scipy.stats.norm.pdf(x,mean3,std)
y6 = scipy.stats.norm.pdf(x,mean6,std)
y7 = scipy.stats.norm.pdf(x,mean6-.5,std_big*1.2)
y4 = scipy.stats.norm.pdf(x,mean1,std_small) * .6
y5 = scipy.stats.norm.pdf(x,mean1-1,std)

meanA = 4.65
meanB = 5.35
meanC = 5.6
A = scipy.stats.norm.pdf(x,meanA,std)
B = scipy.stats.norm.pdf(x,meanB, std)
C = scipy.stats.norm.pdf(x,meanC,std_big)
```

```python
fig, ax = plt.subplots(1,3, figsize=(6.5,2.1))
sns.set_style()
sns.despine(bottom=True, left=True)
y_arr = [y2, y3, y4]
plt.subplots_adjust(hspace=.2, wspace=.1)
titles = ['Consistent', 'Inconsistent', 'Subset']
x_offset = [0, 7.2, 15.8]
for i, a in enumerate(ax):
    a.set_xlim(x_min+1.5,x_max-1.5)
    a.set_ylim(0, 0.6)
    a.set_xticks([])
    a.set_yticks([])
    a.fill_between(x, y_arr[i], color=(1,0,0), alpha=.75, label=r'$\mathcal{T}$')
    if i == 0:
        a.set_ylabel('NE Count', fontsize=12)
    if i == 1:
        a.set_xlabel("Signal parameter", fontsize=12)
        a.fill_between(x, y5, color=(0,0,1), alpha=.66)
    else:
        a.fill_between(x, y1, color=(0,0,1), alpha=.66, label=r'$\mathcal{R}$')
    plt.text(-12+x_offset[i], .54, titles[i], fontsize=10, color=(0,0,0))
    handles, labels = a.get_legend_handles_labels()

plt.legend(handles[::-1], labels[::-1], prop={'size': 7}, frameon=False)
plt.savefig('Consistency_Figures/param_dist.pdf', dpi=400, bbox_inches='tight', format='pdf')
```

```python
fig, ax = plt.subplots(1,1, figsize=(4.5,4))
sns.set_style()
sns.despine(bottom=True, left=True)
ax.set_xlim(x_min,x_max)
ax.set_ylim(.001, 0.3)
ax.set_xticks([])
ax.set_yticks([])
ax.fill_between(x, y3*.65, color=(.6,.2,.2), alpha=.6, label=r'$MM_A$')
ax.fill_between(x, y6*.65, color=(.6, .6, .2), alpha=.6, label=r'$MM_B$')
ax.fill_between(x, y7*.95, color=(.2,.2,.2), alpha=.6, label=r'$MM_C$')
ax.set_xlabel("Signal parameter", fontsize=16)
ax.set_ylabel("NE Count", fontsize=16)
plt.legend(prop={'size':11})
plt.savefig('Consistency_Figures_pdf/param_dist_A_B_C.pdf', dpi=400, bbox_inches='tight', format='pdf')
plt.savefig('Consistency_Figures_png/param_dist_A_B_C', dpi=400, bbox_inches='tight')
```
